# The ‘Hello World’ of neural networks

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
tf.__version__
```

```python
#one layer with one neuron and one output
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
#optimizer and loss function
model.compile(optimizer='sgd', loss='mean_squared_error') #optimizer: stochastic gradient descent
```

```python
#data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)
```

```python
model.fit(xs, ys, epochs = 500)
```

```python
print(model.predict([10.0]))
```

Como hay muy pocos datos, no es posible para la maquina aprender de forma muy precisa la relación entre ellos, por lo que la perdida es alta, y el resultado no es muy bueno, teniendo en cuenta que $y = 2x -1$
